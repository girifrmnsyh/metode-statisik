```python
# import package
import pandas as pd
import numpy as np
```

```python
# import data
data = pd.read_csv('data/data_kos_its.csv', sep=',')
data
```

### Lokasi

```python
data['Lokasi'].unique()
```

```python
# delete 'Kelurahan' in colom 'Lokasi'.

location = []

for d in data['Lokasi']:
    location.append(d.split()[-1])

location
```

```python
# change value
data['Lokasi'] = location
```

```python
data.head()
```

### Fasilitas

```python
# change '·' to ',' and throw the ratings

facilities = []

for d in data['Fasilitas']:
    L = d.split('\n')[0]
    facilities.append(L.replace('·',', '))

facilities[:10]
```

```python
# change value
data['Fasilitas'] = facilities
```

```python
data.head()
```

### Harga

```python
# remove 'Rp' and '.'

prices = []

for d in data['Harga']:
    p = d.replace('Rp','')
    p2 = p.replace('.','')
    prices.append(int(p2))

prices[:5]
```

```python
data['Harga'] = prices

data
```

```python
data.to_csv('data/data_kos_its_clean.csv', index=False)
```
