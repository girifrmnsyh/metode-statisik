```python
# import package
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
import pandas as pd
import time

import warnings
warnings.filterwarnings('ignore')
```

```python
# path set up
PATH = 'C:\chromedriver-win64\chromedriver.exe'
service = Service(executable_path=PATH)
driver = webdriver.Chrome(service=service)

# open page
url = 'https://mamikos.com/kost/kost-dekat-its-surabaya-murah?gender=1&rent=2&sort=price,-&price=10000-3000000&singgahsini=0&tag=15,59'
driver.get(url)
driver.maximize_window()
time.sleep(3)
```

```python
# list for data
names = []
locations = []
facilities  = []
stars = []
prices = []

# collect data
for page in range(10):

    time.sleep(15)
    # all item in a page
    items = driver.find_elements(by='xpath', value='//div[@class="room-list__card col-12"]')
    print(f'Page {page+1} showed.')

    # get data value
    for item in items:

        # names
        try:
            name = item.find_element(by='xpath', value='.//span[@class="rc-info__name bg-c-text bg-c-text--title-4"]').text
            names.append(name)
        except:
            name = item.find_element(by='xpath', value='.//span[@class="rc-info__name bg-c-text bg-c-text--title-4 track-list-booking-kost"]').text
            names.append(name)

        # locations
        try:
            location = item.find_element(by='xpath', value='.//span[@class="rc-info__location bg-c-text bg-c-text--body-1"]').text
            locations.append(location)
        except:
            location = item.find_element(by='xpath', value='.//span[@class="rc-info__location bg-c-text bg-c-text--body-1 track-list-booking-kost"]').text
            locations.append(location)

        # facilities
        try:
            
            facility = item.find_element(by='xpath', value='.//div[@class="rc-facilities rc-facilities--horizontal"]').text
            facilities.append(facility)
        except:
            facility = item.find_element(by='xpath', value='.//div[@class="rc-facilities rc-facilities--horizontal track-list-booking-kost"]').text
            facilities.append(facility)

        # stars
        try:
            try:
                star = item.find_element(by='xpath', value='.//div[@class="rc-facilities__rating-wrapper"]').text
                stars.append(star)
            except:
                stars.append('-')
        except:
            try:
                star = item.find_element(by='xpath', value='.//div[@class="rc-facilities__rating-wrapper track-list-booking-kost"]').text
                stars.append(star)
            except:
                stars.append('-')
        
        # prices
        try:
            price = item.find_element(by='xpath', value='.//span[@class="rc-price__text bg-c-text bg-c-text--title-4"]').text
            prices.append(price)
        except:
            price = item.find_element(by='xpath', value='.//span[@class="rc-price__text bg-c-text bg-c-text--title-4 track-list-booking-kost"]').text
            prices.append(price)

    # click next button
    driver.find_element(by='xpath', value='//a[text()=">"]').click()
    time.sleep(3)
```

```python
# make a DataFrame

result_data = pd.DataFrame({'Nama' : names,
                            'Lokasi' : locations,
                            'Fasilitas' : facilities,
                            'Review' : stars,
                            'Harga' : prices})
result_data.head(10)
```

```python
result_data.to_csv('data/data_kos_its.csv', index=False)
```
